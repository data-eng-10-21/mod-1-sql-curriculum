```python
import pandas as pd
```

```python
data = pd.read_html('https://www.espn.com/nba/stats/player/_/table/offensive/sort/avgPoints/dir/desc')
```

```python
player_names = data[0]
```

```python
player_stats = data[1]
```

```python
player_names.head()
```

```python
player_stats.head()
```

```python
top_fifty = pd.merge(player_names, player_stats, left_index=True, right_index=True)
```

```python
import sqlite3
nba_db = sqlite3.connect('nba_players.db')
```

```python
top_fifty.to_sql('players', nba_db)
```

```python
cursor = nba_db.cursor()
cursor.execute('SELECT * FROM players LIMIT 1')
# cursor.fetchall()
```
